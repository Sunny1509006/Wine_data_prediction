```python
df['fixed acidity']=df['fixed acidity']+0.71
df['volatile acidity']=df['volatile acidity']+0.71
df['citric acid']=df['citric acid']+0.71
df['chlorides']=df['chlorides']+0.71
df['free sulfur dioxide']=df['free sulfur dioxide']+0.71
df['total sulfur dioxide']=df['total sulfur dioxide']+0.71
df['density']=df['density']+0.71
df['sulphates']=df['sulphates']+0.71
df['alcohol']=df['alcohol']+0.71
```

```python
df.head()
```

```python
df['wine quality']=0
for index, item in df.iterrows():
    if item['quality']>5:
        df['wine quality'][index]=2
    else:
        df['wine quality'][index]=1
```

```python
train_data=df
train_data=train_data.drop(columns=['quality'])
train_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(train_data)
scaler.transform(train_data)
```

```python
train_data=train_data.drop(columns=['wine quality'])
train_data.head()
```

```python
from sklearn.model_selection import train_test_split

y = df['wine quality']
X = train_data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
# Create Decision Tree classifer object
DTC = DecisionTreeClassifier()

# Train Decision Tree Classifer
DTC = DTC.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

Y_pred = DTC.predict(X_test)

Y_pred = np.where(Y_pred >=1.5, 2, 1)
print(accuracy_score(y_test, Y_pred))
```

```python
print(Y_pred)
```

```python
#confusion matrix
from sklearn import metrics

confusion_matrix = metrics.confusion_matrix(y_test, Y_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

```python
fpr, tpr, _ = metrics.roc_curve(y_test-1,  Y_pred-1)
auc = metrics.roc_auc_score(y_test-1, Y_pred-1)
#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```

```python
sensitivity = confusion_matrix[0,0]/(confusion_matrix[0,0]+confusion_matrix[0,1])
print('Sensitivity : ', sensitivity )

specificity = confusion_matrix[1,1]/(confusion_matrix[1,0]+confusion_matrix[1,1])
print('Specificity : ', specificity)
```
